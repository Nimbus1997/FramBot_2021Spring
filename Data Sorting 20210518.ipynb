## 1. RGB이미지로 파일 골라내기 (처음부터)

```python
# path= '/home/vision/Documents/Farmbot/LiDAR_20210506/data' #필요없는 것 지울 폴더
# path= './data3_ALLselect' #필요없는 것 지울 폴더
path= './data1_RGBselect' #필요없는 것 지울 폴더
```

```python
#1,2,3 처음부터 모든 폴더 들어가보며 이미지 체크하고 지우기

# 중간에 esc누르면 작업멈추기위해 필요한 변수 선언
exitouterloop=False
start_from_this_dir="" #중간에 멈추면 멈춘폴더 저장


for dirtory in os.listdir(path): # data폴더내 파일(폴더)list
    print("[New Dirtory] : ",dirtory) #현재 폴더 출력
    
        
    for file in os.listdir(path+"/"+dirtory):
         # 1. RGB*.png 파일을 열어서 이미지 sorting 하기
        if "RGB" in file:
            
            img=path+"/"+dirtory+"/"+file # 이미지 파일명 with path
            print(img)
            
            # 2. 이미지를 띄워서 필요한 사진인지 아닌지 판단
            imgg=cv.imread(img, cv.IMREAD_COLOR)
            cv.imshow('press SPACE to delete and ANY OTHER KEY to just pass',imgg)
            k=cv.waitKey(0)
            #(key들어오는 것 대기 시간[ms] 0은 무한) k에는 이제 들어온 key값저장 Ascii코드 보기
            cv.destroyAllWindows() 
            
            
            # 3. 꽃, 식물 끝, 가지 분기점 이 3개가 모두 보이는 경우를 제외하고는 space를 눌러서 제거
            if k==32: #space를 누르면 같은 이름인 것 지우기
                num=file.split(".")[0].split("_")[5]
                [os.remove(f) for f in glob.glob(path+"/"+dirtory+"/*"+num+".*")]
            
            if k==27: #esc를 누루면, 종료하기
                exitouterloop=True
                print("esc")
                break
                
    if exitouterloop:
        start_from_this_dir=dirtory
        print("\n\n>>>>>>",start_from_this_dir)
        break
```

## 6. RGB와 Depth 동시에 확인 (","누르면 rgb다시 볼 수 있음 (한번))

```python
from PIL import Image #image 파일 열기
import os #주소로 폴더 열기
import cv2 as cv
import glob
import shutil

path= './data1_RGBselect' #필요없는 것 지울 폴더
path2="./data2_DEPTHselect"
path3="./data3_ALLselect"
```

```python
#1,2,3 처음부터 모든 폴더 들어가보며 이미지 체크하고 지우기

# 중간에 esc누르면 작업멈추기위해 필요한 변수 선언
# 아스키코드 [ , : 44 . : 46 ]
exitouterloop=False
start_from_this_dir="" #중간에 멈추면 멈춘폴더 저장


for dirtory in os.listdir(path3): # data폴더내 파일(폴더)list
    print("[New Dirtory] : ",dirtory) #현재 폴더 출력
    
        
    for file in os.listdir(path3+"/"+dirtory):
         # 1. RGB*.png 파일을 열어서 이미지 sorting 하기
        if "RGB" in file :
            
            img=path3+"/"+dirtory+"/"+file # 이미지 파일명 with path2
            print(img)
            
            # 2. 이미지를 띄워서 필요한 사진인지 아닌지 판단
            imgg=cv.imread(img, cv.IMREAD_COLOR)
            cv.imshow('press SPACE to delete and ENTER to see Depth(End: ESC)',imgg)
            k=cv.waitKey(0)
            #(key들어오는 것 대기 시간[ms] 0은 무한) k에는 이제 들어온 key값저장 Ascii코드 보기
            cv.destroyAllWindows() 
            
            
            # 3. 꽃, 식물 끝, 가지 분기점 이 3개가 모두 보이는 경우를 제외하고는 space를 눌러서 제거
            if k==32: #space를 누르면 같은 이름인 것 지우기 --RGB만 봐도 지워야할 경우
                num=file.split(".")[0].split("_")[5]
                [os.remove(f) for f in glob.glob(path3+"/"+dirtory+"/*"+num+".*")]
            if k==27: #esc를 누루면, 종료하기
                exitouterloop=True
                print("esc")
                break    
                
            if k ==13: #Enter key를 누르면 같은 이름인 depth 이미지 확인
                name_dept="_".join(file.split("_")[0:3])+"_Depth_"+"_".join(file.split("_")[4:])
                
                imgd=path3+"/"+dirtory+"/"+name_dept # 이미지 파일명 with path3
                print(imgd)
                imggd=cv.imread(imgd, cv.IMREAD_COLOR)
                cv.imshow('press SPACE to delete and ENTER to pass(End: ESC)',imggd)
                k=cv.waitKey(0)
                #(key들어오는 것 대기 시간[ms] 0은 무한) k에는 이제 들어온 key값저장 Ascii코드 보기
                cv.destroyAllWindows() 
                if k ==44 : # ,누루면 (<) RGB이미지 한번더
                    imgg=cv.imread(img, cv.IMREAD_COLOR)
                    cv.imshow('press .',imgg)
                    kk=cv.waitKey(0)
                    cv.destroyAllWindows()
                    if kk ==46:
                        imgd=path3+"/"+dirtory+"/"+name_dept # 이미지 파일명 with path3
                        imggd=cv.imread(imgd, cv.IMREAD_COLOR)
                        cv.imshow('press SPACE to delete and ENTER to pass(End: ESC)',imggd)
                        k=cv.waitKey(0)
                        #(key들어오는 것 대기 시간[ms] 0은 무한) k에는 이제 들어온 key값저장 Ascii코드 보기
                        cv.destroyAllWindows() 
                    
                if k==32: #space를 누르면 같은 이름인 것 지우기 
                    num=file.split(".")[0].split("_")[5]
                    [os.remove(f) for f in glob.glob(path3+"/"+dirtory+"/*"+num+".*")]
                if k==27: #esc를 누루면, 종료하기
                    exitouterloop=True
                    print("esc")
                    break

                
    if exitouterloop:
        start_from_this_dir=dirtory
        print("\n\n>>>>>>",start_from_this_dir)
        break
```
